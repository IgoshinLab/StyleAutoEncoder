```python
import pandas as pd

# --- 1. Define File Paths ---
# Set the locations of your input files.
WORKING_DIR = "/home/xavier/Documents/DAE_project"
split_file = f"{WORKING_DIR}/Tros-classify/classifier_full_dataset.csv"
phenotype_file = f"{WORKING_DIR}/Tros-classify/details.csv"

# --- 2. Load the Data ---
# Read the two CSV files into pandas DataFrames.
print("Loading data...")
split_df = pd.read_csv(split_file)
phenotype_df = pd.read_csv(phenotype_file)

# --- 3. Merge the DataFrames ---
# Combine the two DataFrames based on the 'run_id' and 'scope_id' columns.
print("Merging data...")
merged_df = pd.merge(
    split_df,
    phenotype_df,  # Merging the full dataframe is safer
    on=['run_id', 'scope_id'],
    how='inner'  # 'inner' means we only keep rows that have matching IDs in both files.
)

# --- 4. Keep and select unique rows ---
print("Dropping duplicate rows and selecting columns...")
# First select the specific columns you want, then drop duplicates.
# This ensures we only check for duplicates based on the columns we are keeping.
unique_df = merged_df[["run_id", "scope_id", "phenotype", "Category", "name"]].drop_duplicates()

# --- 5. Rename Columns ---
# The attribute is .columns (plural) and it expects a simple list.
unique_df.columns = ["Run", "Scope", "Phenotype", "Category", "Name"]

# --- 6. Generate LaTeX for Overleaf ---
# This prints the DataFrame in a format that can be copied into a .tex file.
print("\n--- LaTeX Output for Overleaf ---")
# Using the new column names in the LaTeX output
print(unique_df.to_latex(index=False))

# --- 7. Display the Result in Notebook ---
# Show the first 5 rows of your final, unique table.
print("\nMerge and deduplication complete! Here is a preview of your data:")
unique_df

```
